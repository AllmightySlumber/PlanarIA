## DONNÉES

```python
# Import des données de projet
# Il peut être nécessaire d'insaller openpyxl pour pouvoir éxécuter cette cellule
planaire=pd.read_excel("datalocomotion.xlsx") # on importe le data set depuis Seaborn
```

On va venir préparer les données.
> Séparation des variables(X), des étiquettes (Y) des variables(X)

Pour notre base de données actuel l'étiquette correspond au traitement utilisé, et les variables correspondent au colonnes d1,...,d10.

On va venir choisir quelle donné va servir d'étiquette et à partir de quelles autres données on essaie de le trouver (i.e lorsqu'on classifie on cherche à deviner y à partir de quoi (x)?).

```python
X = planaire[["d1","d2","d3","d4","d5","d6","d7","d8","d9","d10",]]
X.head() # Affichage des premiers éléments de X (nos variables)
```

```python
y = planaire["traitement"]
y.head()
```

Il est toujours intéressant d'afficher les dimenssions de nos bases de données.

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.20, random_state=2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## KNN

```python
# On définit le modèle (knn) avec sklearn en choisissant
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier(n_neighbors= 2) # on considère les 3 plus proches voisins
```

#### APPRENTISSAGE / VALIDATION

Comme il est possible de le voir sur l'image ci-dessous, on a d'abord commencé par considérer les 3 plus proches voisins. Avec ce nombre de voisin, le score du modèle était de 42,5%. On à penser que ce n'était pas suffisant. Donc on a joué avec ce paramètre pour essayer de trouver la valeur pour laquelle le score est le plus élevé.

<img src="knn_3.png" alt="score" width="800" height="700">

On a finalement trouver que le score est le plus élevé lorsque qu'on considère seulement les 2 plus proches voisins.

```python
# Entrainement du model sur les données d'entrainement
knn.fit(X_train, y_train)
```

Faisons nous une idée des performances du modèle en affichant la learning_curve.

```python
# Définition des tailles d'échantillon pour la learning curve
from sklearn import model_selection
train_sizes = np.linspace(0.1, 1.0, 10)  # 10 tailles entre 10% et 100% des données

# Calcul des scores d’apprentissage et de validation
train_sizes, train_scores, val_scores = model_selection.learning_curve(
    knn, X_train, y_train, cv=5, train_sizes=train_sizes, scoring='accuracy', n_jobs=-1
)

# Moyenne et écart-type des scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
val_mean = np.mean(val_scores, axis=1)
val_std = np.std(val_scores, axis=1)

# Tracé de la courbe
plt.figure(figsize=(8, 6))
plt.plot(train_sizes, train_mean, 'o-', color='blue', label='Score entraînement')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color='blue', alpha=0.2)

plt.plot(train_sizes, val_mean, 'o-', color='red', label='Score validation')
plt.fill_between(train_sizes, val_mean - val_std, val_mean + val_std, color='red', alpha=0.2)

# Personnalisation du graphique
plt.xlabel("Taille de l'échantillon d'entraînement")
plt.ylabel("Score de précision")
plt.title("Score de précision selon la taille des données d'entrainement")
plt.legend()
plt.grid(True)
plt.show()
# La taille du jeu de données vous semble-t-elle suffisante?
```

On peut voir que le score d'entrainement, qui est le score lorsqu'on test le modèle sur les mêmes données qui ont servi à son entrainement, est assez faible comparé à ce qu'on attendrait. On peut alors supposé que:
> le modèle a du mal a apprendre les données d'entrainement.  

Le fait que le score de validation soit beaucoup plus faible que le score d'entrainement nous permet d'émettre cette hypothèse.
> le modèle à du mal à généralisé et fait du surapprentissage

#### ANALYSE DES PERFORMANCES

Voyons maintenant les performances réelles du modèle sur les données de test.

```python
knn.score(X_test, y_test)
```

Sans surprise le score du modèle n'est pas très élevé. Voyons avec la matrice de confusions quelles sont les étiquettes qu'il confond.

```python
y_pred = knn.predict(X_test)
cm = sklearn.metrics.confusion_matrix(y_test, y_pred)
```

```python
# Afficher cette matrice de confusion à l'aide de la fonction heatmap de seaborn
plt.figure(figsize=(6, 5))
sb.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=set(y_test), yticklabels=set(y_test))
plt.xlabel("Prédictions")
plt.ylabel("Étiquette attendue")
plt.title("Matrice de confusion du KNN")
plt.show()
```

On peut voir que le modèle a prédit à tort:
> 6 fois l'étiquette 4 comme étant 0  
>  4 fois l'étiquette 0 comme étant 2  
>  3 fois l'étiquette 2 comme étant 0  

Les étiquettes que le modèle reconnait le mieux reconnus sont le 0 et le 3 avec 7 prédictions correcte pour les deux.

Ce qu'il serait intéressant de faire maintenant serait de voir si on combinant les modèles moins performantes on pourrait obtenir en combinant leurs prédictions un modèles plus performant qui pourrait éventuellement être comparer au réseaux de neurones.  
En observant un peut et en comparant les matrices de confusion, on a remarqué que les modèles ne confondent pas les mêmes étiquettes et pas avec la même fréquence. L'idée serait d'accorder un poid à leurs avis selon l'étiquette qu'il reconnaissent le mieux individuellement et lorsque qu'il votent l'étiquette avec le plus de poid sera celle prédit.

## RÉSEAUX DE NEURONES À LA MAIN

Dans cette dernière partie le but est de réaliser un réseaux de neurones à 2 couche avec 2 neurones. La fonction d'activation que nous allons utiliser est la sigmoïd du fait de la simplicité des calculs que ça va engendrer lors de l'étape de la régression. Pour notre fonction coût on utilisera l'erreur quadratique moyenne.  
On va construire les fonctions séparément, les tester ensuite on les assemblera pour faire le réseaux de neurones.

#### LES FONCTIONS

On va d'abord commencer par la fonction d'activation (sigmoïd) qui est assez simple.  
<img src="sigmoid.png" alt="score" width="500" height="400">

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def test_sigmoïd():
    assert sigmoid(0)==0.5, "Échec test 1"
    assert sigmoid(1)== (1 / (1 + np.exp(-1))), "Échec test 2"
    assert sigmoid(-1)== (1 / (1 + np.exp(1))), "Échec test 3"
    print("Tous les tests ont réussi !")

test_sigmoïd()
```

Ensuite la fonction d'erreur quadratique moyenne.  

<img src="mse.png" alt="score" width="500" height="400">

```python
""" y_att est le tableau des valeurs attendu et y_pred est le tableau de ce qui à été prédit. """
def mse(y_att, y_pred):
    total_error = 0
    n = len(y_att)
    
    for i in range(n):
        total_error += (y_att[i] - y_pred[i]) ** 2
    
    return total_error / n
```

```python
def test_mse():
    y_true = np.array([1, 0, 1])
    y_pred = np.array([1, 0, 1])
    assert mse(y_true, y_pred) == 0, "Échec test 1"
    
    y_true = np.array([1, 2, 3])
    y_pred = np.array([2, 2, 2])  # Différence de 1, 0 et -1
    
    assert mse(y_true, y_pred)==2/3, "Échec test 2"
    
    print("Tous les tests ont réussi !")

test_mse()
```

Il reste encore à faire une fonction de propagation avant, une fonction pour la descente de gradient avant de pouvoir tout assemblé dans une fonction et obtenir notre réseaux de neurones.

```python
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1
    A1 = sigmoid(Z1)

    Z2 = np.dot(A1, W2) + b2
    A2 = sigmoid(Z2)

    return A1, A2
```

```python
""" Fonction pour calculer la descente de gradient et renvoyer les nouvelles valeurs pour les paramètres
"""
def backward_propagation(X, Y, A1, A2, W2, learning_rate):
    m = X.shape[0]

    dZ2 = (A2 - Y) * A2 * (1 - A2)  # Dérivée de la sigmoïde
    dW2 = np.dot(A1.T, dZ2) / m
    db2 = np.sum(dZ2, axis=0, keepdims=True) / m

    dZ1 = np.dot(dZ2, W2.T) * A1 * (1 - A1)
    dW1 = np.dot(X.T, dZ1) / m
    db1 = np.sum(dZ1, axis=0, keepdims=True) / m

    return dW1, db1, dW2, db2
```

La fonction tracer_courbe ci-dessous va nous servir pour pouvoir tracer la courbe d'apprentissage de notre réseaux et donc de savoir si il a bien appris.

```python
# Fonction pour tracer la courbe d'apprentissage
def tracer_courbe(loss_history, epochs=1000):
    plt.plot(range(epochs), loss_history, label="Loss")
    plt.xlabel("Époques")
    plt.ylabel("Erreur quadratique moyenne (MSE)")
    plt.title("Courbe d'apprentissage")
    plt.legend()
    plt.grid()
    plt.show()
```

Maintenant que on a toute les parties qui vont composer notre réseaux on peut les assembler. Pour ça, on crée une fonction neural_network avec les paramètres d'entrées X Y loss_history qui va servir de tableau pour stocker l'erreur quadratique moyenne au cours des itérations, epochs qui représente le nombre d'itération max (1000 par défaut) et la learning_rate (0.01 par défaut).

```python
def neural_network(X, Y, loss_history, epochs=1000, learning_rate=0.01):
    input_dim = X.shape[1]
    hidden_dim = 2
    output_dim = Y.shape[1] if Y.ndim > 1 else 1  # Pour vérifier la forme de Y (s'il est unidimensionnel)

    # Initialisation des poids aléatoirement
    W1 = np.random.randn(input_dim, hidden_dim) * 0.01
    b1 = np.zeros((1, hidden_dim))
    W2 = np.random.randn(hidden_dim, output_dim) * 0.01
    b2 = np.zeros((1, output_dim))

    for epoch in range(epochs):
        # Propagation avant
        A1, A2 = forward_propagation(X, W1, b1, W2, b2)

        # Calcul de l'erreur
        loss = mean_squared_error(Y, A2)
        loss_history.append(loss)  # Stocker la perte

        # Backpropagation
        dW1, db1, dW2, db2 = backward_propagation(X, Y, A1, A2, W2, learning_rate)

        # Mise à jour des poids
        W1 -= learning_rate * dW1
        b1 -= learning_rate * db1
        W2 -= learning_rate * dW2
        b2 -= learning_rate * db2

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Loss: {loss:.4f}")

    return W1, b1, W2, b2
```

Il nous faut également une fonction predict qui permet de tester le réseaux sur les données de test.

```python
# Fonction de prédiction sur les données de test
def predict(X, W1, b1, W2, b2):
    _, A2 = forward_propagation(X, W1, b1, W2, b2)
    # Retourner les indices des classes avec la probabilité la plus élevée
    return np.argmax(A2, axis=1)  # A2 représente la sortie de la couche de sortie
```
